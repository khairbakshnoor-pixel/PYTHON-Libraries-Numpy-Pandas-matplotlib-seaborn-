```python
# Add column BMI = Weight / Height**2
df['BMI']=df['Weight'] / (df['Height']**2)
```

```python
# Add column Class:

# Heavy if Weight > 100

# Medium if Weight 50‚Äì100

# Light if Weight < 50
def classify(weight):
    if weight > 100:
        return 'Heavy'
    elif weight >= 50:
        return 'Medium'
    else:
        return 'Light'

# Apply the function to the Weight column
df['Class'] = df['Weight'].apply(classify)
```

```python
# Add column Fast = True if Height > 2
df['Fast']=df['Height']>2
```
