```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fuelstations.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Federal Agency ID','Federal Agency Name','Updated At','Intersection Directions',
                 'Plus4','Station Phone','Geocode Status','Groups With Access Code'], inplace=True)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Street Address'] = df['Street Address'].fillna('Unknown')
```

```python
df.info()
```

```python
# df['Street Address'] = df['Street Address'].fillna('Unknown')

for col in df.select_dtypes(include='object').columns:
  df[col] = df[col].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
for col in df.select_dtypes(include=['float64','int64']).columns:
  df[col] = df[col].fillna(df[col].median())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
fuel_mix = df['Fuel Type Code'].value_counts()
fuel_mix
```

```python
fuel_mix.plot(kind='bar', title='Energy Product Portfolio Distribution')
plt.show()
```

Above insight has been concluded due to the reason that multi-energy portfolios reduce downstream risk

```python

df['Open Date']= pd.to_datetime(df['Open Date'], errors='coerce')
df.info()
```

```python
df['Open Date'].head()
```

```python
open_year = df['Open Date'].dt.year.value_counts().sort_index()
open_year.plot(figsize=(10,5),title='Energy Insfrastructure Expansion over Time')
plt.show()
```

```python
city_counts = df['City'].value_counts().head(15)

city_counts.plot(kind='bar', title='Stations count by City')
plt.show()
```

```python
ev_df = df[df['Fuel Type Code'].str.contains('ELEC')]
ev_state = ev_df['State'].value_counts().head(10)
ev_state.plot(kind='bar', title='Electric Vehicle Stations by State')
plt.show()
```
