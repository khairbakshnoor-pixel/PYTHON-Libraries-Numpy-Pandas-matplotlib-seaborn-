```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x=[2022,2023,2024,2025]
y=[15,20,25,3]
z=[23,56,45,34]
plt.plot(x,y,
         marker='o',
         alpha=0.5,
         color='red',
         markersize=30,
         markerfacecolor='black',
         markeredgecolor='blue',
         linestyle='--',
         linewidth=4)
plt.title("result",fontweight='bold')
plt.plot(x,z,
         marker='o',
         alpha=0.5,
         color='red',
         markersize=15,
         markerfacecolor='black',
         markeredgecolor='blue',
         linestyle='--',
         )
plt.grid(color='yellow',
         linestyle='--')
plt.xlabel("years",color='green',
           size=15)
plt.ylabel('sales ',
           color='blue',
           size=15)
plt.xticks(color='blue')
plt.yticks(fontweight='bold',color='gray')
plt.legend()
plt.show()
```

```python
#bar chats
category=np.array(['vegtable','meat','milk','rice'])
value=np.array([2,5,8,3])
plt.bar(category,value,color='red')
plt.title("Eating time table",fontweight='bold',size=20,color='darkgreen')
plt.xlabel("eating things",size=15,color='blue')
plt.ylabel("quantity",size=15,color='blue')
plt.xticks(color='brown')
plt.yticks(color='brown')
plt.show()
```

```python
#HORIZANTAL BAR GRAPH
category=np.array(['vegtable','meat','milk','rice'])
value=np.array([2,5,8,3])
plt.barh(category,value,color='red')
plt.title("Eating time table",fontweight='bold',size=20,color='darkgreen')
plt.xlabel("eating things",size=15,color='blue')
plt.ylabel("quantity",size=15,color='blue')
plt.xticks(color='brown')
plt.yticks(color='brown')
plt.show()
```

```python
#pie charts
category=np.array(['fresher','senior','teachers','pm'])
value=np.array([200,225,300,120])
colors=np.array(['red','blue','yellow','green'])
plt.pie(value,labels=category,autopct='%.1f %%',colors=colors,explode=[0.1,0.1,0.1,0.1],shadow=True)
plt.title("szabist university",fontweight='bold',color='brown')

plt.show()
```

```python
#scatter graph
x=[1,2,3,4,5,6,7,8]
y=[10,20,30,20,40,50,70,10]
x1=[10,20,30,20,40,50,70,10]

y1=[1,2,3,4,5,6,60,70]
plt.scatter(x,y,color='red',alpha=0.7,s=200,label="A")
plt.scatter(x1,y1,color='blue',alpha=0.7,s=200,label="B")
plt.title("Test scores")
plt.legend()
plt.xlabel("hour studied",family='algerian')
plt.ylabel("marks",family='algerian')
plt.show()
```

```python
#histogram
scores=np.random.normal(loc=60,scale=20,size=100)
scores=np.clip(scores,0,100)
plt.hist(scores,bins=20,color='green',edgecolor='black')
plt.title("exam scores")
plt.xlabel("scores")
plt.ylabel("Number of students")
plt.show()
```
