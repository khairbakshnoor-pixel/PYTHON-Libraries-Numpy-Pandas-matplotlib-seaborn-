```python
Q1 = df['Daily_Phone_Hours'].quantile(0.25)
Q3 = df['Daily_Phone_Hours'].quantile(0.75)
IQR = Q3 - Q1
LF= Q1- (1.5 * IQR)
UF= Q3+ (1.5 * IQR)

outliers = df[(df['Daily_Phone_Hours'] < LF) | (df['Daily_Phone_Hours'] > UF)]
outliers
```
