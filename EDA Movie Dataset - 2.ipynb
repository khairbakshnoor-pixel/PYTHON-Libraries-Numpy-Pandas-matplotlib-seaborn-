```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('movie.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
df.info()
```

```python
df['Runtime'] = df['Runtime'].str.replace('min','')
```

```python
df['Runtime'].head()
```

```python
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')
```

```python
df.info()
```

```python
df['Gross'].head()
```

```python
df['Gross'] = pd.to_numeric(df['Gross'],errors='coerce')
```

```python
df['Released_Year'] = pd.to_numeric(df['Released_Year'],errors='coerce')
```

```python
df.info()
```

```python
df['Certificate'].head()
```

```python
df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].median())
df['Gross'] = df['Gross'].fillna(df['Gross'].median())
df['Released_Year'] = df['Released_Year'].fillna(df['Released_Year'].median())
```

```python
df['Certificate'] = df['Certificate'].fillna(df['Certificate'].mode()[0])
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['IMDB_Rating'], bins=20)
plt.title('IMDB Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Gross'], bins=30)
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```

```python
df.describe()
```

```python
top_gross = df.sort_values(by='Gross',ascending=False).head(10)
print(top_gross[['Series_Title','Gross']])
```

```python
#rating vs Revenue Relationship
plt.figure(figsize=(8,5))
plt.scatter(df['IMDB_Rating'],df['Gross'])
plt.xlabel('IMDB Rating')
plt.ylabel('Revenue')
plt.title('Rating vs Revenue')
plt.show()
```

```python
#Runtime DIstribution
#Genre Revenue Analysis (groupby)
#Director Performance
#Votes vs Rating
#Revenue by Year
```

```python
#Runtime Distribution
plt.figure(figsize=(8,5))
plt.hist(df['Runtime'], bins=20)
plt.xlabel('Minutes')
plt.ylabel('Frequency')
plt.show()
```

```python
#Genre Revenue Analysis (groupby)
df['Genre']
```

```python
df['Primary_Genre'] = df['Genre'].str.split(',').str[0]
```

```python
df['Primary_Genre']
```

```python
genre_revenue = df.groupby('Primary_Genre')['Gross'].mean().sort_values(ascending=False)
genre_revenue
```

```python
genre_revenue.plot(kind='bar', figsize=(8,5), title='Average Revenue by Genre')
plt.ylabel('Average Revenue')
plt.show()
```

```python
#Director Performance
director_revenue = df.groupby('Director')['Gross'].mean().sort_values(ascending=False).head(10)
director_revenue
```

```python
director_revenue.plot(kind='bar', figsize=(3,3), title='Top directors by avg revenue')
plt.ylabel('Revenue')
plt.show()
```

```python
#Votes vs Rating
plt.figure(figsize=(7,7))
plt.scatter(df['No_of_Votes'], df['IMDB_Rating'])
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.title('Votes vs Rating')
plt.show()
```

```python
#Avg Revenue by year
yearly_revenue = df.groupby('Released_Year')['Gross'].mean()
yearly_revenue
```

```python
yearly_revenue.plot(title='Avg Revenue by Year')
plt.ylabel('Average Gross')
plt.show()
```
