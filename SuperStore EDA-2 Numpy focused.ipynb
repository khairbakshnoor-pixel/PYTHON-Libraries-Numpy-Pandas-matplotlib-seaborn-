```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Superstore.csv', encoding='latin1')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

```python
df.head()
```

```python
df.info()
```

```python
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df['Profit Margin'] = df['Profit']/df['Sales']
```

```python
#Filter rows where sales is > 1000
df[df['Sales']>1000]
```

```python
#Filter rows with loss
df[df['Profit']<0]
```

```python
#Calculate total sales per category
df.groupby('Category')['Sales'].sum()
```

```python
#Total profit per region
df.groupby('Region')['Profit'].sum()
```

```python
#Average discount per segment
```

```python
#top 10 highest sales
df.sort_values(by='Sales', ascending=False).head(10)
#top 10 biggest losses
#plot histogram of sales
# plot bar chart of sals by category
# line plot of yearly sales
yearly_sales = df.groupby('Year')['Sales'].mean()
yearly_sales.plot(kind='line')
plt.show()
```
