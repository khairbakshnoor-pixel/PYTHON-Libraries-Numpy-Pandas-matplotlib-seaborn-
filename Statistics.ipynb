Types of Data:

1. Qualitative (Categorical data):
1a. Nominal: Gender, Color, Country
1b. Ordinal: Categorical data in an specific order like
Ranks, Grades, Levels
2. Quantitative(Numerical data):
2a. Discrete:Count data (students, cars, siblings)
2b. Continuous:Height, Weight, temperature

Statistics

Central Tendency
Mean (Average):
[5,7,9,11,13] => Mean:9
[2,4,6,8,90] => Mean:22
Mean is sensitive to extreme values

Median: Middle value of sorted data
[5,7,9,11,13] => Median:9
[2,4,6,8,90] => Median:6

Mode: Most repeated / frequent value

M,M,M,F,F,M,F,F,F,M,F => Mode: F

Outlier: Extreme values in your data that disturb the mean of the data.
