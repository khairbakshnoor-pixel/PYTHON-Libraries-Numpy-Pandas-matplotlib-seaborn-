Types of Data:

1. Qualitative (Categorical data):
1a. Nominal: Gender, Color, Country
1b. Ordinal: Categorical data in an specific order like
Ranks, Grades, Levels
2. Quantitative(Numerical data):
2a. Discrete:Count data (students, cars, siblings)
2b. Continuous:Height, Weight, temperature

Statistics

Central Tendency
Mean (Average):
[5,7,9,11,13] => Mean:9
[2,4,6,8,90] => Mean:22
Mean is sensitive to extreme values

Median: Middle value of sorted data
[5,7,9,11,13] => Median:9
[2,4,6,8,90] => Median:6

Mode: Most repeated / frequent value

M,M,M,F,F,M,F,F,F,M,F => Mode: F

Outlier: Extreme values in your data that disturb the mean of the data.

```python
import numpy as np
x = [10,20,30,40,50]
print('Mean:',np.mean(x))
print('Median:',np.median(x))
```

```python
#Percentiles in python using numpy
q1 = np.percentile(x, 25)
q2 = np.percentile(x, 50)
q3 = np.percentile(x, 75)
print(q1,q2,q3)
```

```python
IQR=q3-q1
IQR
```

```python
UF=q1 + 1.5*(IQR)
UF
#Uf is the upper fence
```

```python
LF=q1-1.5*(IQR)
LF
#lf is the lower fence
```

```python
array=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
array
```

```python
q1 = np.percentile(array, 25)
q2 = np.percentile(array, 50)
q3 = np.percentile(array, 75)
print(q1,q2,q3)
```

```python
IQR=q3-q1
IQR
```

```python
UF=q1 + 1.5*(IQR)
UF
```

```python
LF=q1-1.5*(IQR)
LF
```

```python
l=[]
for i in array:
    if UF>=i and LF<=i:
        l.append(i)
arr2=np.array(l)
arr2
```
