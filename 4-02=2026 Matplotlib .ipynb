```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Basic line plot
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y,
         color='b' #line color
         , linestyle=':',
         linewidth='1.5', marker='o',
         markersize=8, markerfacecolor='green',
         markeredgecolor='black', label='sample line', alpha=0.8)
plt.title('Line Plot Demo', fontsize=20)
plt.xlabel('x-axis', fontsize=14)
plt.ylabel('y-axis', fontsize=14)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=12)
plt.legend()
plt.grid(True, linestyle=':', linewidth=1, alpha=0.7)

plt.savefig('plot.png')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
plt.scatter(
    x,y,
    c='green',
    s=100, #size
    alpha = 0.7, #transparency
    marker = '^',
    edgecolors='black'
)
plt.show()
```

```python
#Bar chart
categories=['A','B','C','D']
values=[10,25,15,30]

plt.bar(
    categories, values,
    color=['red','blue','green','orange'],
    width=0.6,
    edgecolor='black'
)
plt.show()
```

```python
plt.barh(categories, values, color='purple')
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30,
         color='skyblue',
         edgecolor='black',
         alpha=0.8)
plt.show()
```

```python
x = np.arange(0, 10, 0.1)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.show()
```
