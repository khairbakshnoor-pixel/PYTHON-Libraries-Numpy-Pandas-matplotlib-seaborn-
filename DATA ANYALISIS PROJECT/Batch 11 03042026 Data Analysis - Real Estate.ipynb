```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
a = df.groupby('NoConv')['Price'].mean()
a
```

```python
# Avg price based on the num of convenience stores
a = df.groupby('NoConv')['Price'].mean()
plt.bar(x= a.index,height=a.values)
plt.title('Avg price based on no. of conv stores')
plt.xlabel('No. of Stores')
plt.ylabel('Price')
plt.show()
```

```python
#New Feature: Price per store ratio
df['PPSR'] = df['Price'] / (df['NoConv'] + 1)
```
